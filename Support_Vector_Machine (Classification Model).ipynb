```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.05, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Creating Model --> SVM (Support Vector Machine)

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel = 'linear', random_state = 0)
```

```python
classifier.fit(x_train, y_train)
```

# Predictions

```python
print(classifier.predict(sc.transform([[32, 150000]]))[0])
```

```python
print(classifier.predict(sc.transform([[20, 86000]]))[0])
```

```python
y_prediction = classifier.predict(x_test)
```

```python
y_prediction
```

```python
print(np.concatenate((y_prediction.reshape(len(y_prediction), 1), y_test.reshape(len(y_test), 1)), 1)) # [Prediction, Actual]
```

## Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_prediction) # See in Google... [correct, wrong] ... We can use confusion matrix to evaluate our model.
cm                                                           # [wrong, correct]
```

```python
accuracy_score(y_test, y_prediction) # Accuracy of 90% is achieved.
```
